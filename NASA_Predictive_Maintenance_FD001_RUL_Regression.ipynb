But we can also let the feature engineering be done by the neural network itself. So we keep these sensor values for this notebook.

```python
# function to reshape features into (samples, time steps, features) 
def sequence_from_df(id_df, seq_length, seq_cols):
    """ Only sequences that meet the window-length are considered, no padding is used. This means for testing
    we need to drop those which are below the window-length. An alternative would be to pad sequences so that
    we can use shorter ones """
    data_array = id_df[seq_cols].values
    num_elements = data_array.shape[0]
    # print(f'Num Elements: {num_elements}') # helped in find error
    # for row index start, row index end in
    for start, stop in zip(range(0, num_elements-seq_length), range(seq_length, num_elements)):
        # return the data array containing the current 50 rows from the original sequence.
        yield data_array[start:stop, :]
```

```python
def df_to_seq_array(df, seq_function, seq_len):
    # select the features for the model
    #sensor_cols = ['s' + str(i) for i in range(1,22)]
    # these are the feature columns
    sequence_cols = ['id', 's11', 's12', 's13', 's14', 's15', 's17', 's2', 's20',
       's21', 's3', 's4', 's7', 's8', 's9', 'setting1', 'setting2'] 
    #sequence_cols.extend(sensor_cols)
    
    # get seq data from df
    seq_gen = (list(seq_function(df[df['id']==id_], seq_len, sequence_cols))
           for id_ in df['id'].unique())
    
    # prepare array for model
    seq_gen = [seq for seq  in seq_gen if len(seq) != 0] # timeseries of length less that 50 in test set causes empty array
    seq_array = np.concatenate(list(seq_gen)).astype(np.float32)
    
    return seq_array
    
    
    
```

```python
test_df.head()
```

```python
train_X = df_to_seq_array(train_df, seq_function=sequence_from_df, seq_len=sequence_length)
test_X = df_to_seq_array(test_df, seq_function=sequence_from_df, seq_len=sequence_length)
```

```python
print(train_X.shape, test_X.shape)
```

```python
# function to generate labels
def gen_labels_for_id(id_df, seq_length, label):
    data_array = id_df[label].values
    num_elements = data_array.shape[0]
    return data_array[seq_length:num_elements, :]
```

```python
# Function to generate the target variable for the train and test data
def get_target_labels(df, get_label_func, seq_len, target_label):
    label = [target_label]
    id_labels = [get_label_func(df[df['id']==id_], sequence_length, label) 
             for id_ in df['id'].unique()]
    label_array = np.concatenate(id_labels).astype(np.float32)
    return label_array
    
    
    
```

```python
train_y = get_target_labels(train_df, gen_labels_for_id, sequence_length, 'RUL')
test_y = get_target_labels(test_df, gen_labels_for_id, sequence_length, 'RUL')
print(train_y.shape, test_y.shape)
```

#### We also need to scale the target values

```python
target_scaler = preprocessing.MinMaxScaler()
target_scaler.fit(train_y)
```

```python
# transform target variables
train_y = target_scaler.transform(train_y)
test_y = target_scaler.transform(test_y)
```

## LSTM Model for Classification

As mentioned earlier, we will use a sequence of length 50.

### We will use Weights and Biases API for logging the the model performance for multipls runs



```python
import wandb
import os
from wandb.keras import WandbCallback

```

```python
# Set hyperparameters, which can be overwritten with a W&B Sweep
hyperparameter_defaults = dict(
    dropout = 0.2,
    layer_1_size = 48,
    layer_2_size = 24,
    output_neuron = 1,
    batch_size = 512,
    epochs = 50,
)

# Initialize wandb
wandb.init(project="nasa-predictive-maintenance", config=hyperparameter_defaults)
config = wandb.config
```

```python


# build the network
nb_features = train_X.shape[2]
nb_out = train_y.shape[1]

model = Sequential()
model.add(LSTM(units=config.layer_1_size, input_shape=(sequence_length, nb_features),  return_sequences=True))
model.add(Dropout(config.dropout))
model.add(LSTM(units=config.layer_2_size, return_sequences=False))
model.add(Dropout(config.dropout))
model.add(Dense(units=config.output_neuron, activation='linear'))
model.compile(loss='mse', optimizer='adam')
print(model.summary())


```

```python
%%time
# fit the network
# callbacks = [keras.callbacks.EarlyStopping(monitor='val_loss', min_delta=0, patience=0, verbose=0, mode='auto')]
history = model.fit(train_X, train_y, 
                    epochs=config.epochs, 
                    batch_size=config.batch_size, 
                    validation_data=[test_X, test_y], 
                    verbose=1,
                    callbacks = [WandbCallback()]
          )


# Save model to wandb
model.save(os.path.join(wandb.run.dir, "model.h5"))
```

### Vizualising the Accuracy and the Loss of the trained model.

```python
plt.figure(figsize=(8,5))

plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label= 'test')
plt.legend()
plt.title('Loss Plot')
plt.xlabel('Training Epochs')
plt.ylabel('Loss')

plt.show()
```

### Evaluating model on the last sequence in the test data

```python
# Function to get the last sequence in each time series in test data
sequence_cols = ['id', 's11', 's12', 's13', 's14', 's15', 's17', 's2', 's20',
       's21', 's3', 's4', 's7', 's8', 's9', 'setting1', 'setting2']
def get_last_seq(df,seq_cols, target_label, seq_len = 50):
    ids = df['id'].unique()
    X_list = []
    y_list = []
    
    for i in ids:
        if(len(df[df['id']== i]) >=50):
            temp = df[df['id']== i][seq_cols].values[-seq_len:]
            X_list.append(temp)
            y_list.append(df[df['id']== i][target_label].values[-1])
    X_array = np.asarray(X_list).astype(np.float32)
    y_array = np.asarray(y_list).astype(np.float32)
    y_array = y_array.reshape(y_array.shape[0],1)
    
    return X_array, y_array
```

```python
%%time
last_X, last_y = get_last_seq(test_df, sequence_cols, 'RUL', seq_len=50)
print(last_X.shape, last_y.shape)
```

```python
y_actual = last_y # last_y is unscaled.
y_pred = model.predict(last_X, batch_size=2048, verbose=0)
#print(y_actual[0], y_pred[0])


from sklearn.metrics import mean_squared_error
from math import sqrt

# y_actual is unscaled, hence we need to inverse transform prediction before calculating the rmse
rms = sqrt(mean_squared_error(y_actual, target_scaler.inverse_transform(y_pred)))
print(f"RMSE on the last cycle of each series is: {rms}")
```
